```python
# Maximum Non-Overlapping Meetings (Greedy)

def max_meetings(meetings):

    meetings.sort()

    count = 1
    last_end = meetings[0][0]

    for i in range(1, len(meetings)):
        start = meetings[i][1]
        end = meetings[i][0]

        if start >= last_end:
            count += 1
            last_end = end

    return count



T = int(input())

for t in range(T):
    N = int(input())
    meetings = []

    for i in range(N):
        S, E = map(int, input().split())
        meetings.append((E, S))

    print(max_meetings(meetings))
```

```python
def first_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            high = mid - 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans


def last_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            low = mid + 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans


T = int(input())
for _ in range(T):
    N, Q = map(int, input().split())
    arr = list(map(int, input().split()))

    query_values = list(map(int, input().split()))
    for x in query_values:
        first = first_occurrence(arr, x)
        if first == -1:
            print(0)
        else:
            last = last_occurrence(arr, x)
            print(last - first + 1)
```
