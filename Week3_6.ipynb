```python
T = int(input())

for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))
    BIT = [0] * (N + 1)
    def update(i, diff):
        i += 1
        while i <= N:
            BIT[i] += diff
            i += i & -i
    def prefix_sum(i):
        i += 1
        s = 0
        while i > 0:
            s += BIT[i]
            i -= i & -i
        return s

    for i in range(N):
        update(i, arr[i])

    Q = int(input())

    for _ in range(Q):
        query = list(map(int, input().split()))

        if query[0] == 1:
            i, x = query[1], query[2]
            diff = x - arr[i]
            arr[i] = x
            update(i, diff)
        else:
            L, R = query[1], query[2]
            print(prefix_sum(R) - prefix_sum(L - 1))
```
