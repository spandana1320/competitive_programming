**Problem - 1 :**

```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    def update(self, i, delta):
        while i <= self.n:
            self.tree[i] += delta
            i += i & -i

    def query(self, i):
        total = 0
        while i > 0:
            total += self.tree[i]
            i -= i & -i
        return total

print("\n--- Student Attendance Analysis ---")

# INPUT SECTION
n = int(input("Enter total number of days: "))

arr = list(map(int, input(
    "Enter attendance points for each day (space-separated): ").split()))

ft = FenwickTree(n)

# Build tree
for i in range(n):
    ft.update(i + 1, arr[i])

# Query
day_query = int(input(
    "Enter the day number to calculate total attendance up to that day: "))

print("Total attendance till Day", day_query, "=", ft.query(day_query))

# Update
day_update = int(input(
    "Enter the day number where attendance needs correction: "))

new_value = int(input(
    "Enter the NEW attendance value for that day: "))

delta = new_value - arr[day_update - 1]
arr[day_update - 1] = new_value
ft.update(day_update, delta)

# Query again
day_query = int(input(
    "Enter the day number to calculate attendance after update: "))

print("Updated total attendance till Day", day_query, "=", ft.query(day_query))
```
